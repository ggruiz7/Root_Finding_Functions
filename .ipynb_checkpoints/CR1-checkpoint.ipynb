## The Bisection Method

```python
def f(x):
    return 0.5 * math.sin(x-1)**3

def g(x):
    return 6**(-x) - 2

list = []

def Bisection(f, a, b, tol): 
    condition = True
    
    while condition == True:
        # check if a and b bound a root
        if np.sign(f(a)) == np.sign(f(b)):
            print('a and b DO NOT bound a root of f')
            condition = False

        # midpoint
        p = (a + b) / 2 

        # stopping condition
        if np.abs(f(p)) < tol:
            list.append(p)
            print('\nThe root of the function between these bounds is ', p)
            condition = False
        # f(a)*f(p) > 0, change a to p
        elif np.sign(f(a)) == np.sign(f(p)):
            list.append(p)
            print('Approximation = %f' % (p))
            return Bisection(f, p, b, tol)
        # change b to p
        elif np.sign(f(b)) == np.sign(f(p)):
            list.append(p)
            print('Approximation = %f' % (p))
            return Bisection(f, a, p, tol)
        
        
    
Bisection(f, -1, 2, 0.000000001)

list2 = []

for i in list:
    err = abs(i - 1.0)
    list2.append(err)

Bisection(g, -4, 4, 0.00000001)

list3 = []

for i in list:
    err = abs(i - (-0.38685280829668045))
    list3.append(err)

plt.plot(range(len(list2)), list2)
plt.plot(range(len(list3)), list3)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title('Error of Bisection Method of f1 and f2')
plt.legend(['f1', 'f2'])
plt.show()
```

## Fixed Point Iteration

```python
def f(x):
    return 0.5 * math.sin(x-1)**3
# rewrting f(x):
def fomega(x):
    return x - 3.5 * (0.5 * math.sin(x-1)**3)

def g(x):
    return 6**(-x) - 2
# rewriting g
def gomega(x):
    return x + 1/3 * (6**(-x) - 2)

list = []

def Fixed_Point(f, x0, tol, maxiter):
    iteration = 1
    condition = True
    while condition == True:
        x1 = f(x0)
        print('Iteration: %d, Approximation = %f' % (iteration, x1))
        x0 = x1
        
        list.append(x0)

        iteration = iteration + 1
        
        if iteration > maxiter:
            print('\nNot Convergent in 100 iterations.')
            break
        
        if abs(f(x0)) < tol:
            print('\nApproximation: %f' % (x1))
        
        condition = abs(f(x0)) > tol
        if condition == False:
            break
            
        
    return x1

Fixed_Point(fomega, -2, 0.00000001, 100)

list4 = []

for i in list:
    err = abs(i - 1.0)
    list4.append(err)

Fixed_Point(gomega, 0, 0.00000001, 100)

list5 = []

for i in list:
    err = abs(i - (-0.38685280829668045))
    list5.append(err)

plt.plot(range(len(list4)), list4)
plt.plot(range(len(list5)), list5)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title('Error of Fixed Point Method on f1 and f2')
plt.legend(['f1', 'f2'])
plt.show()
```

## Newton's Method

```python
def f(x):
    return 0.5 * math.sin(x-1)**3
def fprime(x):
    return 1.5 * math.sin(x-1)**2 * math.cos(x-1)

def g(x):
    return 6**(-x) - 2
def gprime(x):
    return -math.log(6) / 6**x

list = []

def Newton(f, fprime, x0, tol):
    x1 = x0 - f(x0) / fprime(x0)
    
    if abs(f(x0)) < tol:
        print('\nApproximation: %f' % (x0))
        list.append(x0)
        return x0
    else:
        print('Approximation: %f' % (x0))
        list.append(x0)
        return Newton(f, fprime, x1, tol)

Newton(f, fprime, 2, 0.00000001)

list6 = []

for i in list:
    err = abs(i - 1.0)
    list6.append(err)

Newton(g, gprime, -2, 0.00000001)

list7 = []
    
for i in list:
    err = abs(i - (-0.38685280829668045))
    list7.append(err)
    
plt.plot(range(len(list6)), list6)
plt.plot(range(len(list7)), list7)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title("Error of Newton's Method on f1 and f2")
plt.legend(['f1', 'f2'])
plt.show()
```
