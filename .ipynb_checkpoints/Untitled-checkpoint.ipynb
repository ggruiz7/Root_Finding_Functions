## Our functions f & g and their derivatives

```python
import sympy as sym

x = sp.Symbol('x')
f = (1/2)*(sym.sin(x-1))**3
print('f(x) =', f)
fprime = f.diff(x)
print("f'(x) =", fprime)


x = sp.Symbol('x')
g = 6**(-x) - 2
print('g(x) =', g)
gprime = g.diff(x)
print("g'(x) =", gprime)
```

## Fixed Point Iteration

```python
def f(x):
    return 0.5 * math.sin(x-1)**3
# rewrting f(x):
def fomega(x):
    return 0.5 * math.sin(x-1)**3 + 1

def g(x):
    return 6**(-x) - 2
# rewriting g
def gomega(x):
    return -math.log(x + 2) / math.log(6)

list = []

def Fixed_Point(f, x0, tol, maxiter):
    iteration = 1
    condition = True
    while condition == True:
        x1 = f(x0)
        print('Iteration: %d, Approximation = %f' % (iteration, x1))
        x0 = x1
        
        list.append(x0)

        iteration = iteration + 1
        
        if iteration > maxiter:
            print('\nNot Convergent in 100 iterations.')
            break
        
        if abs(f(x0)) < tol:
            print('\nApproximation: %f' % (x1))
        
        condition = abs(f(x0)-x0) > tol
        
    return x1

Fixed_Point(fomega, 5, 0.00000001, 100)

list2 = []

for i in list:
    err = abs(i - 1.0)
    list2.append(err)

Fixed_Point(gomega, -1, 0.00000001, 100)

list3 = []

for i in list:
    err = abs(i - (-0.38685280829668045))
    list3.append(err)

plt.plot(range(len(list2)), list2)
plt.plot(range(len(list3)), list3)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title('Error of Fixed Point Method on f and g')
plt.legend(['f', 'g'])
plt.show()
```

## Newton's Method

```python
def f(x):
    return 0.5 * math.sin(x-1)**3
def fprime(x):
    return 1.5 * math.sin(x-1)**2 * math.cos(x-1)

def g(x):
    return 6**(-x) - 2
def gprime(x):
    return -math.log(6) / 6**x

list = []

def Newton(f, fprime, x0, tol):
    x1 = x0 - f(x0) / fprime(x0)
    
    if abs(f(x0)) < tol:
        print('\nApproximation: %f' % (x0))
        list.append(x0)
        return x0
    else:
        print('Approximation: %f' % (x0))
        list.append(x0)
        return Newton(f, fprime, x1, tol)

Newton(f, fprime, 2, 0.00000001)

list2 = []

for i in list:
    err = abs(i - 1.0)
    list2.append(err)

Newton(g, gprime, -2, 0.00000001)

list3 = []
    
for i in list:
    err = abs(i - (-0.38685280829668045))
    list3.append(err)
    
plt.plot(range(len(list2)), list2)
plt.plot(range(len(list3)), list3)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title("Error of Newton's Method on f and g")
plt.legend(['f', 'g'])
plt.show()
```

## The Secant Method

```python
def f(x):
    return 0.5 * math.sin(x-1)**3

def g(x):
    return 6**(-x) - 2

list = []

def Secant(f, x0, x1, tol, maxiter):
    iteration = 1
    condition = True
    while condition == True:
        if f(x0) == f(x1):
            print('Error: dividing by 0 (f(x1)-f(x0) = 0)')
            break
            
        
        x2 = x0 - (x1-x0)*f(x0) / (f(x1)-f(x0)) 
        print('Iteration: %d, Approximation = %f' % (iteration, x2))
        
        list.append(x2)
        
        x0 = x1
        x1 = x2
        iteration = iteration + 1
        
        if iteration > maxiter:
            print('Not Convergent in 100 iterations.')
            break
        
        if abs(f(x2)) < tol:
            print('\n Approximation: %f' % x2)
            
        condition = abs(f(x2)) > tol
    
Secant(f, 0, 3, 0.00000001, 100)

list2 = []

for i in list:
    err = abs(i - 1.0)
    list2.append(err)

Secant(g, -2, 2, 0.00000001, 100)

list3 = []
    
for i in list:
    err = abs(i - (-0.38685280829668045))
    list3.append(err)
    
print(list2)
print(list3)
    
plt.plot(range(len(list2)), list2)
plt.plot(range(len(list3)), list3)
plt.xlabel('Number of Iterations')
plt.ylabel('Error')
plt.title("Error of Newton's Method on f and g")
plt.legend(['f', 'g'])
plt.show()
```
